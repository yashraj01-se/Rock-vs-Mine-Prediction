Importing the Dependencies


```python
import numpy as np #used to handle arrays...
import pandas as pd #used to handle several data processing step...upload data into data frames...
from sklearn.model_selection import train_test_split #used to split the dataset into traning and testing data,so that we do not have to do it manually...
from sklearn.linear_model import LogisticRegression #logistic regression model function....
from sklearn.metrics import accuracy_score #used to find out accuracy of our model...
```

Data Collection and Data Processing

```python
#Loading the Dataset to a pandas Dataframe
sonar_data=pd.read_csv('/content/sonar data.csv',header=None) #header=None (is used when there are no header in our dataset(column header))
```

```python
#used ton display first five row of our Dataset...
sonar_data.head()
```

```python
# used to display rows and columns...
sonar_data.shape
```

```python
#used to give us the Statistical data...
sonar_data.describe() #25% means 25 percent of values is less than 0.13350....
```

```python
#used to display the number of different bases like in this project(rocks and mines)....
sonar_data[60].value_counts() #[60] to represent in which coulmn is those bases stored in our dataset...(Remainder:number of rocks and mines rows should be close to each other for correct prediction..)
```

M->Mine

R->Rock


```python
#grouping of dataset...
sonar_data.groupby(60).mean() #dufference between M and R is important in terms of mean...
```

```python
#used to seperate Data and labels..(Supervised Learning)...
X=sonar_data.drop(columns=60,axis=1) #used to drop a particular column or row from the Dataset...use axis=0 in case of row...
Y=sonar_data[60]
```

```python
print(X)
print(Y) #sucessfully splited the Data and Label...
```

Splitting Data into training and testing Data


```python
#order should be followed-first traning data,testing data then labels of traning and testing Data ...
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1) #test_size=0.1 means that we want 0.1 means 10% of dataset to be testing Data..Stratify is jused to split Dtaset correctly...random_state is used to split dataset in particular order...
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model Training->Logistic Regression

```python
print(X_train)
print(Y_train)
```

```python
model=LogisticRegression()
```

```python
#training the Logistic Regression model using Traning Data...
model.fit(X_train,Y_train) #we are using data and label of traning data for traning of the model.....
```

Model Evaluation

```python
#used to find out accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train) #Y_train is actual answer...
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
#used to find out accuracy on testing data
X_test_prediction=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test) #Y_test is actual answer...

```

```python
print('Accuracy of testing data : ',testing_data_accuracy)
```
